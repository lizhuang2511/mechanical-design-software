```python
#-*-coding=utf-8-*-
#@time:2024/9/711:36
#Author:lizhuang
#@File:吊耳计算.py
#@Software:PyCharm
#定义设计参数
import xlwings

t=20#吊耳板厚度，单位：mm
d=44#销孔直径，单位：mm
a=45#外轮廓半径，单位：mm
H=100#销孔高度，单位：mm
B=200#吊耳宽度，单位：mm
r=d/2#板孔半径，单位：mm
P=5#设计荷载，单位：吨
e=60#吊耳孔拉力角度，单位：度
k=1.10#动载系数
k2=1.65#安全系数
#钢板材质Q235的屈服强度，这里假设为235N/mm²，但实际应用中需根据材料证明或规范确定
Q=235#单位：N/mm²，注意此处单位转换，原文中Q235mm应为误写，应为Q235N/mm²，这里简化为Q=235

import math

#计算吊耳孔拉力
FL=P/math.sin(math.radians(e))#单位：吨

#计算紧接承压应力
cj=k*FL*9.8*1000/(d*t)#单位：MPa

#计算R值
R=min(B/2,r+a)#单位：mm

#计算孔壁抗拉应力
ck=cj*(R**2+r**2)/(R**2-r**2)#单位：MPa

#计算焊缝受拉荷载
FLO=k*k2*P/math.sin(math.radians(e))#单位：吨

#计算焊缝受水平剪力
Fh=FLO*math.cos(math.radians(e))#单位：吨

#计算焊缝受到弯矩
M=Fh*H/1000#单位：t*m，注意此处单位转换，原文中可能为笔误，应为N*m，但考虑到FLO和Fh的单位为吨，这里保持t*m作为结果单位

#计算焊缝面积
Ae=B*t#单位：mm²

#计算焊缝受到的拉应力
N=(k*k2*P*1000*9.8)/Ae#单位：MPa，注意此处单位转换，将吨转换为N，mm²保持不变

#计算焊缝受到的弯曲应力
#注意：原文中的We未明确定义，但根据公式推测可能为焊缝的截面模量，这里按照矩形截面模量公式计算We=(B*t^2)/6（单位：mm³）
We=(B*t**2)/6#单位：mm³
OM=(6*M*1000*9.8000*1000)/(t*B**2)#单位：MPa，注意此处单位转换，将t*m转换为N*m，并保持MPa作为结果单位
#但由于We的计算可能存在误解，且原文中未给出明确We的定义和计算方法，因此此处的OM计算可能不准确，需根据实际情况调整

#考虑到OM计算的不确定性，这里仅给出N和Ty的计算结果作为焊缝应力的参考
#计算焊缝受到的剪应力
Ty=(Fh*1000*9.8)/Ae#单位：MPa，注意此处单位转换，将吨转换为N，并保持MPa作为结果单位
xgm=math.sqrt((N+OM)**2+(3*Ty**2))
#输出计算结果（这里仅输出部分关键结果作为示例）
print(f"紧接承压应力σcj:{cj:.2f}MPa,许用值80mpa")
print(f"孔壁抗拉应力σck:{ck:.2f}MPa,许用值115mpa")
print(f"焊缝受到的拉应力N:{N:.2f}MPa")
print(f"焊缝受到的剪应力Ty:{Ty:.2f}MPa")
print(f"焊缝受到的弯曲应力om:{OM:.2f}MPa")
print(f"焊缝受到的合成应力Ty:{xgm:.2f}MPa,许用值225mpa")

#注意：原文中的焊缝组合应力计算公式存在错误，无法直接转换为有效的Python代码，
#因此这里没有给出焊缝组合应力的计算结果。
#在实际应用中，需要根据正确的焊缝组合应力计算公式进行计算。
```
