```python
时，两量棒间的内侧距离，见图 $5.3-8 \mathrm{~b}$ 。
a)

![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-1.jpg?height=595&width=1075&top_left_y=1106&top_left_x=1416)
b)

![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-1.jpg?height=1714&width=1305&top_left_y=1692&top_left_x=1337)

图 5.3-8 花键测量

a）外花键持目径向路动 b）内花链棒间距

11）跨棒距 借助两量棒测量外花键实际齿厚时两贯棒间的外侧距㐫，图 5.3-9。

![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-1.jpg?height=1670&width=1554&top_left_y=4085&top_left_x=1141)

图 5.3-9 外花镂等棒距

12) 变换系数 跨棒距值的变换系数, 其值为路棒距的变动最与齿厚的变动量之比。

13）公法线长度 相隔 $K$ 个齿的两外侧齿面各与两平行平面之中的一个平面相切, 此两平面之间的垂直距郖。

14）公法线平均长度 同一花键上实际测得的公法线长度的平均值。

## 2.2 花煡连接的强度计算

### 2.2.1 通用简单算法

此法适用于矩形花键和渐开线花键。

花键连接的类型和尺寸通常根据被连接件的结构特点, 使用要求和工作条件选择。为避免键齿工作表面压溃 (静连接) 或过度廜损 (动连接)，应进行必要的强度校校计算，计算公式觧连接

$$
\sigma_{\mathrm{p}}=\frac{2 T}{\psi Z h l d_{\mathrm{m}}} \leqslant\left[\sigma_{\mathrm{p}}\right]
$$

动连接

$$
p=\frac{2 T}{\psi Z h l d_{\mathrm{m}}} \leqslant[p]
$$

式中 $T$ 一一传递转矩 $(\mathrm{N} \cdot \mathrm{mm})$;

$\psi$ 一一各齿间载荷不均匀系数，一般取 $\psi=0.7$ $\sim 0.8$, 类数多时取偏小值;

$Z$ 一一花键的齿数;

l一齿的工作长度 $(\mathrm{mm})$;

$h$ 一一键工作高度 $(\mathrm{mm})$;

$d_{\infty}$ 一平均直径 $(\mathrm{mm})$;

矩形花键 $h=\frac{D-d}{2}-2 C ; D_{\mathrm{m}}=\frac{D+d}{2}$

浙开线花键 $\quad h=\left\{\begin{array}{l}m \quad \alpha_{D}=30^{\circ} \\ 0.8 m \quad \alpha_{D}=45^{\circ} ;\end{array} \quad D_{\mathrm{m}}=D ;\right.$

$C$ 一一倒角尺寸 ( $\mathrm{mm})$;

$m$ 一一模数 ( $\mathrm{mm})$;

$\left[\sigma_{p}\right]$ 一一花键连接许用接压应力 $(\mathrm{MPa})$, 见表 5. 3-11;

[p]一许用压强 $(\mathrm{MPa})$ ，见表 5.3-11。

表 5.3-11 花链達接的许用掊压应力，许用陇强

| 连接工 <br> 作方式 |  | 许用值 | 使用和制 <br> 造掅况 | 鮆面末经 <br> 热处理 | 䒨面经 <br> 热处理 |
| :---: | :---: | :---: | :---: | :---: | :---: |
| 竫连接 |  | 许用接 <br> E应力 | 不良 <br> 中筞 | $35 \sim 50$ <br> $60 \sim 100$ | $40 \sim 70$ <br> $100 \sim 140$ |
| 动 <br> 连 <br> 接 | 空载下 <br> 移动 | 许用压强 <br> $[p]$ | 不良 <br> 中等 <br> 良好 | $15 \sim 20$ <br> $20 \sim 30$ <br> $25 \sim 40$ | $20 \sim 35$ <br> $30 \sim 60$ <br> $40 \sim 70$ |
|  | 㘺荷作 <br> 用下 <br> 移动 | 许用压强 <br> $[p]$ | 不食 <br> 中等 <br> 良好 | - <br> - <br> - | $3 \sim 10$ <br> $5 \sim 15$ <br> $10 \sim 20$ |

( $\mathrm{MPa})$

注：1. 使用和制造不良，系指受变载，有双向冲击、振动教率高和振幅大，洞清不好（对动连接）、材料硬度不高和精度不高等。

2. 同一情况下, $\left[\sigma_{p}\right]$ 或 $[p]$ 的较小值用于工作时间长和较重要的场合。
3. 内、外花键材料的抗拉强度不低于 $590 \mathrm{MPa}$ 。

## 2. 2.2 矩形花键承载能力计算（精确算法）

GB/T 17855-1999《花键承载能力计算方法》规定了矩形花键承载能力计算的主要内容，包括：花键受载分析、系数的确定和梅面接触强度、齿根抗弯
强度、些根抗前强度、齿面耐唐损能力的计筫方法及外花键扭转与弯曲承载能力计算方法等内容。

（1）常见的失效形式（表 5.3-12）

（2）承载能力计算

在产品设计时, 应根据花键零件的具体结构、受力状态、材料热处理及硬度、精度等级等情况, 选择上述内容的全部或部分进行花键承裁能力计算。
1）术语与代号 在花键承载能力计算中采用的术语和代号见表 5.3-13。

2) 受力分析

a）无载荷 对于无误差的花键连接，在其无载荷状态时 (不计自重, 下同), 内、外花键各齿的中心线（或对称面）是重合的。键貲两侧间隌相等,均为作用侧隙之半，见图 5.3-10a。

## 表 5.3-12 花组常见的失效形式

| 失效形式 | 主要特征 | 主要原因 | 预防措施 |
| :---: | :---: | :---: | :---: |
| 律先面 <br> 压清 | ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-2.jpg?height=433&width=1728&top_left_y=2418&top_left_x=880) | ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-2.jpg?height=442&width=1545&top_left_y=2409&top_left_x=2607) | ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-2.jpg?height=433&width=1765&top_left_y=2409&top_left_x=4132) |
| ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-2.jpg?height=663&width=506&top_left_y=2831&top_left_x=384) | ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-2.jpg?height=663&width=1728&top_left_y=2831&top_left_x=880) | ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-2.jpg?height=663&width=1545&top_left_y=2831&top_left_x=2607) | ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-2.jpg?height=663&width=1774&top_left_y=2831&top_left_x=4132) |
| 锥齿 面 <br> 来伤（冷 <br> 作硕化伤） | ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-2.jpg?height=433&width=1728&top_left_y=3474&top_left_x=880) | ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-2.jpg?height=442&width=1545&top_left_y=3465&top_left_x=2607) | ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-2.jpg?height=433&width=1765&top_left_y=3465&top_left_x=4141) |
| ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-2.jpg?height=533&width=507&top_left_y=3879&top_left_x=393) | ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-2.jpg?height=533&width=1728&top_left_y=3879&top_left_x=880) | ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-2.jpg?height=533&width=1545&top_left_y=3879&top_left_x=2607) | ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-2.jpg?height=533&width=1765&top_left_y=3879&top_left_x=4141) |
| ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-2.jpg?height=773&width=507&top_left_y=4411&top_left_x=393) | ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-2.jpg?height=773&width=1728&top_left_y=4411&top_left_x=880) | ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-2.jpg?height=773&width=1545&top_left_y=4411&top_left_x=2607) | ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-2.jpg?height=773&width=1774&top_left_y=4411&top_left_x=4141) |

表 5.3-13 术语、代号及说明（GB/T 17855-1999）

| 席号 | 术语 | 代量 | 单位 | 说 |
| :---: | :---: | :---: | :---: | :---: |
| 1 | 输人转矩 | $T$ | $\mathrm{~N} \cdot \mathrm{m}$ | 输人算花钺副的转矩 |
| 2 | 输人功就 | $\boldsymbol{P}$ | $k W$ | 辁人给花铭副的功率 |
| 3 | 转速 | $n$ | $\mathrm{r} / \mathrm{min}$ | 花键副的转速 |
| 4 | 名义切向力 | $F_{1}$ | $\mathbf{N}$ | 花致副所受的名义切向力 |
| 5 | 平均圆直垔 | $\overline{d_{\mathrm{m}}}$ | min | 矩形花键大径与小径之和的一半 |
| 6 | 单位稢荷 | $W$ | $\mathrm{~N} / \mathrm{mm}$ | 单一镭齿在单位长度上所受的法向载荷 |
| 7 | 蚸数（姆数） | $N$ | - | 花键的锥数（先数） |
| 8 | 结合长度 | $l$ | $\mathbf{m m}$ | 内花键与外花键相配合部分的长度（按名义值） |
| 9 | 压轴力 | $\bar{F}$ | $\mathrm{~N}$ | 花能别所受的与轴线垂直的径向作用力 |
| 10 | 弯矩 | $M_{\mathrm{b}}$ | $\mathbf{N} \cdot \mathbf{m}$ | 作用在花锋副上的弯矩 |
| 11 | 使用系数 | $\boldsymbol{K}_{\mathbf{1}}$ | - | 主要考虫由于传动系统外部因菜而产生的动力过载影哰的系数 |
| 12 | 齿侧间落系数 | $\boldsymbol{K}_{2}$ | - | ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-2.jpg?height=323&width=3473&top_left_y=7030&top_left_x=2455) |
| 13 | 分配系数 | $\boldsymbol{K}_{3}$ | - | ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-2.jpg?height=148&width=3473&top_left_y=7343&top_left_x=2455) |
| 14 | 轴向偳载系数 | $\boldsymbol{K}_{4}$ | - | ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-2.jpg?height=314&width=3473&top_left_y=7471&top_left_x=2455) |
| 15 | 枈面压应力 | $\overline{\sigma_{H}}$ | $\overline{\mathrm{MPa}}$ | 键齿表面计算的平均接螌压应力 |
| 16 | 工作齿高 | $h_{*}$ | $\mathrm{~mm}$ | 键齿工作高度, $h_{w}=h_{\text {min }}$ |
| 17 | 外花银大径 | $D$ | $\mathrm{~mm}$ | 外花键大径的基本尺寸 |
| 18 | 内花驖小径 | $\bar{d}$ | $\mathrm{~mm}$ | 内花键小径的基本尺寸 |
| 19 | 拻面接触强度 <br> 的计真安全系数 | $S_{\mathrm{H}}$ | - | $S_{\mathrm{H}}$ 值一没可取 $1.25-1.50$ 。 <br> 较重要的及淬火的花健取较大值，一般的未经淬火的花键取较小值 |


| 庚号 | 术语 | 代号 | 单位 | 说 |
| :---: | :---: | :---: | :---: | :---: |
| 20 | 茜面许用压应力 | $\left[\sigma_{\mathrm{H}}\right]$ | $\mathrm{MPa}$ |  |
| 21 | 材料的屈胆强度 | $\sigma_{0.2}$ | $\mathrm{MPa}$ | 花键材料的屈腿强度（按表层取值） |
| 22 | 啮根弯曲应力 | $\sigma_{5}$ | $\mathbf{M P a}$ | 花键齿根的计算弯曲应力 |
| 23 | 全齿高 | $h$ | $\mathbf{m m}$ | 花煡的全啮高, $h=(D-d) / 2$ |
| 24 | 弦䛈厚 | $S_{\mathrm{Pa}}$ | $\mathrm{mm}$ | ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-3.jpg?height=166&width=2684&top_left_y=1796&top_left_x=2781) |
| 25 | 许用齿根弯曲应力 | $\left[\sigma_{F}\right]$ | $\mathrm{MPa}$ |  |
| 26 | 材料的抗拉强度 | $\sigma_{b}$ | $\mathrm{MPa}$ | 花键材料的抗拉强度 |
| 27 | 抗弯强度的 <br> 计算安全系数 | $S_{F}$ | - | 一般情况 $S_{\mathrm{F}}$ 取 $1.25 \sim 2.00$ |
| 28 | 齿根最大切应力 | $\tau_{\text {Fmax }}$ | $\mathrm{MPa}$ |  |
| 29 | 切应力 | $\tau_{\text {bn }}$ | $\mathrm{MPa}$ | 筷近花键收尾处的第应力 |
| 30 | 应力集中系数 | $\alpha_{\text {ln }}$ | 二 |  |
| 31 | 外花键小径 | $d$ | $\mathrm{~mm}$ | 外花键小径的基本尺寸 |
| 32 | 作用直径 | $d_{\mathrm{h}}$ | mm | 当量应力处的直径, 相当于光滑扭棒的直径 |
| 33 | 齿根圆角半经 | $\rho$ | $\mathrm{mm}$ | 一般指外花链齿根国现最小曲率半径 |
| 34 | 许用切应力 | $\left[\tau_{F}\right]$ | $\mathrm{MPa}$ |  |
| 35 | 婍面察损许用压应力 | $\left[\sigma_{H 1}\right]$ | $\mathbf{M P a}$ | 花键副在 $10^{8}$ 次循环数以下工作时的许用压应力 |
| 36 | 枈面箱损许用医应力 | $\left[\sigma_{\mathrm{H} 2}\right]$ | $\mathrm{MPa}$ | 花锥副长期工作无㻤损的许用压应力 |
| 37 | 当䩗应力 | $\sigma_{v}$ | $\mathrm{MPa}$ | 计算花键扭转与抗弯强度时，切应力与弯曲应力的合成应力 |
| 38 | 弯曲应力 | $\sigma_{\mathbf{F}_{\mathbf{n}}}$ | $\mathrm{MPa}$ | 计算花篗扭转与抗弯强度时的弯曲应力 |
| 39 | 转换系数 | $\boldsymbol{K}$ | 一 | 确定作用䁅径 $d_{\mathrm{b}}$ 的转换系数 |
| 40 | 许用应力 | $\left[\sigma_{v}\right]$ | $\mathrm{MPa}$ | 计算花键扭转与抗弯强度时的许用应力 |
| 41 | 作用驸袕 | $C_{\mathrm{v}}$ | $\mathrm{mma}$ | 花键副的全齿㥜陌 |
| 42 | 位移䨘 | $e_{0}$ | $\operatorname{mon}$ | 花健副的内外花链两轴线的径向相对位移量 |

b) 受纯转矩载荷 对无误差的花链连接, 在其只传递转矩 $T$ 而无压轴力 $F$ 时, 同侧的各齿面在转矩的作用下, 彼此接触、作用侧照相等, 内、外花链的两轴线仍是同轴的, 见图 5.3-10b。所有链齿承受同样大小的载荷, 见图 5.3-11。

c) 受纯压轴力载荷 对无误差的花键连接，在只承受压轴力 $F$ 不受转矩 $T$ 时, 内、外花的两轴线出现一个相对位移量 $e_{0}$ (见图 5.3-10c)。当花链副回转时, 各键齿两侧面所受载荷的大小按图 5.3-12 周期性变化。此时, 花键副容易睹损。

d）受转矩和压轴力两种载荷对无误差的花键连接, 在其承受转矩 $T$ 和压轴力 $F$ 两种载荷时，内、外花链的相对位置和各键齿所受载荷的大小和方向, 决定于所受转矩 $T$ 和压轴力 $F$的大小及两者的比例。

当花键副所受的裁荷主要是转矩 $T$, 压轴力 $F$ 是次要的或很小时，该花键副回转后，各键先两侧面的受力状态发生周期性变化，见图 5.3-13。

当花键副所受的载荷主要是压轴力 $F$ ，转矩 $T$ 是次要的或很小时，该花键副回转后，各键裚两侧面受力状态发生周性变化，见图 5.3-14。在这种情况下，花键副也容易屡损。

![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-3.jpg?height=1272&width=1184&top_left_y=6744&top_left_x=829)
a)

![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-3.jpg?height=1288&width=1172&top_left_y=6745&top_left_x=2611)
b)

![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-3.jpg?height=1235&width=1358&top_left_y=6736&top_left_x=4605)
c)

图 5.3-10 内、外渐开线花傁的相对位管

a）无裁荷、有间暴 b) 只承受转矩 $T$ 无压轴力 $F \quad$ c) 只承受压轴力 $F$ 无转矩 $T$
![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-4.jpg?height=1146&width=2638&top_left_y=901&top_left_x=426)

图 5.3-11 只传递转矩 $T$ 无压轴力 $F$ 时的载荷分配

![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-4.jpg?height=1225&width=2637&top_left_y=2478&top_left_x=422)

图 5.3-12 只承受压轴力 $F$ 而无转矩 $T$ 时的载荷分配

![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-4.jpg?height=1172&width=2633&top_left_y=4121&top_left_x=424)

图 5.3-13 同时承受转矩 $T$ 和压轴力 $F$,转矩 $T$ 占优势时的載荷分配
![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-4.jpg?height=1136&width=2628&top_left_y=5746&top_left_x=426)

图 5.3-14 同时承受压轴力 $F$ 和转矩 $T$,而压轴力占优势时的䧕荷分配

3）花键承载能力计算中的系数

a) 使用系数 $K_{1}$ 使用系数 $K_{1}$ 主要考虑由于传动系统外部因素引起的动力过载影响的系数。

该系数可以通过精密测量获得, 也可经过对全系统分析后确定。在上述方法不能实现时, 可参考表 5. 3-14 取值。

b) 齿侧间隙系数 $K_{2}$ 当花键副承受压轴力 $F$ 作用时, 其各键齿的受力状态将失去均匀性。因花键侧隙发生变化, 内、外花键的两轴线将出现一个位移量 $e_{0}$, 见图 5.3-10c。该位移量会影响花键的承戟能力。这一影响用婍间傹系数 $K_{2}$ 予以考虑。对小径定心的矩形花键, 可取 $K_{2}=1.1 \sim 2.0$ 。

象 5.3-14 使用系数 $K_{1}$

| 原动机 <br> (辖人端) | 工作机（输出端） |  |  |
| :---: | :---: | :---: | :---: |
|  | 中等冲击 | 严重冲击 |  |
| 轻微冲击 | 1.00 | 1.25 | 1.75 或更大 |
| 中筞冲击 | 1.25 | 1.50 | 2.00 或更大 |

注: 1. 均为平稳的原动机: 电动机、㑟汽机、燃气轮机等;

2. 轻貒冲击的原动机：多每内燃机等;
3. 中等冲击的原动机：单箖内矮机等;
4. 均分平稳的工作机：电动机、带式输送机、通风机、透平压缩机、均匀密度材料搅拌机等;
5. 中等冲击的工作机：机床主传动、非均匀密度材料

![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-4.jpg?height=142&width=1874&top_left_y=3632&top_left_x=3805)

6. 严重冲击的. $\mathrm{T}$ 作机：冲床、新床、轨机、钻机等。

当压轴力较小、花键副精度较高时, 可取 $K_{2}=1.1$ $\sim 3.0$; 当压轴力较大、花键副精度较低时, 可取 $K_{2}=$ $2.0 \sim 3.0$; 当压轴力为零时 (只承受转矩) , $K_{2}=1.0$ 。

c) 分配系数 $K_{3}$ 花键副的内花键和外花键的两轴线在同轴状态下, 由于花键位置度误差（键齿等分度误差、对称度误差）的影响，使各链齿所受载荷不同。这种影响用分配系数 $K_{3}$ 予以考虑。

符合 GB/T 1144 标准规定的精密传动用的矩形花键， $K_{3}=1.1 \sim 1.2$; 符合该标准规定的一般用途的矩形花键, $K_{3}=1.3 \sim 1.6$ 。对于经过磨合, 各键齿均可参与工作, 且受载荷基本相同的花键副, 取 $K_{3}=1.0$ 。

d) 轴向偏载系数 $K_{4}$ 由于花键侧面对轴线的平行度误差、安装后的同轴度误差和受载后的扭转变形, 使键将沿轴向所受载荷不均匀。用轴向偏裁系数 $K_{4}$ 予以考虑。其值可从表 5.3-15 中选取。

对磨合后的花键副，各锥齿沿轴向载荷分布基本相同时，可取 $K_{4}=1.0$ 。

当花键精度较高、花键结合长度 $l$ 和平均圆直径 $d_{\mathrm{m}}$ 较小时，表 5.3-15 中的轴向偏较系数 $K_{4}$ 取较小值, 反之取较大值。

4）计算公式 矩形花键承载能力计算公式见表 5. 3-16。

例1 中系列矩形花键副: $6 \times 21 \frac{\mathrm{H} 7}{\mathrm{f} 7} \times 25 \frac{\mathrm{H} 10}{\mathrm{a} 11} \times$ $5 \frac{\mathrm{H} 11}{\mathrm{~d} 10}$, 已知输人功率 $P=8.83 \mathrm{~kW}$, 转速 $n=1275 \mathrm{r} /$ $\min$, 输人端连接离合器 (平稳)，输出端连接断轮 (轻微冲击), 花键结合长度 $l=30 \mathrm{~mm}$, 工作将高 $h_{x}$
$=2 \mathrm{~mm}$ 全齿高 $h=2 \mathrm{~mm}$ 先根圆角半烃 $\rho=0.2 \mathrm{~mm}$ 大径、 $D=25 \mathrm{~mm}$ 、小径 $d=21 \mathrm{~mm}$ ，材料为低碳合金钢、表面渗碳淬火、表面硬度为 $58 \sim 64 \mathrm{HRC} 、 \sigma_{0.2} \geqslant$ $965 \mathrm{MPa} 、 \sigma_{\mathrm{b}}=1080 \mathrm{MPa}_{\text {。 }}$ 校核花键强度。

表5.3-15 轴向熖转系数 $K_{4}$

| 系列或模 <br> 数/ $/ \mathrm{mm}$ | 平均四直 <br> 䄱 $d_{\mathrm{m}} / \mathrm{mm}$ | $l / d_{\infty}$ |  |  |
| :---: | :---: | :---: | :---: | :---: |
|  |  | $\leqslant 1.0$ | $>1.0 \sim 1.5$ | $>1.5 \sim 2.0$ |
| 轻系列 <br> 或 $m \leqslant 2$ | $\leqslant 30$ | $1.1 \sim 1.3$ | $1.2 \sim 1.6$ | $1.3 \sim 1.7$ |
|  | $>30 \sim 50$ | $1.2 \sim 1.5$ | $1.4 \sim 2.0$ | $1.5-2.3$ |
|  | $>50 \sim 80$ | $1.3-1.7$ | $1.6 \sim 2.4$ | $1.7 \sim 2.9$ |
|  | $>80-120$ | $1.4 \sim 1.9$ | $1.8 \sim 2.8$ | $1.9 \sim 3.5$ |
|  | $>120$ | $1.5 \sim 2.1$ | $2.0 \sim 3.2$ | $2.1 \sim 4.1$ |
| 中系列或 <br> $2<m \leqslant 5$ | $\leqslant 30$ | $1.2 \sim 1.6$ | $1.3 \sim 2.1$ | $1.4 \sim 2.4$ |
|  | $>30-50$ | $1.3 \sim 1.8$ | $1.5 \sim 2.5$ | $1.6 \sim 3.0$ |
|  | $>50 \sim 80$ | $1.4 \sim 2.0$ | $1.7 \sim 2.9$ | $1.8-3.6$ |
|  | $>80 \sim 120$ | $1.5-2.2$ | $1.9 \sim 3.3$ | $2.0 \sim 4.2$ |
|  | $>120$ | $1.6-2.4$ | $2.1 \sim 3.6$ | $2.2 \sim 4.8$ |
| $5<m \leqslant 10$ | $\leqslant 30$ | $1.3 \sim 2.0$ | $1.4 \sim 2.8$ | $1.5 \sim 3.4$ |
|  | $>30 \sim 50$ | $1.4 \sim 2.2$ | $1.6 \sim 3.2$ | $1.7 \sim 4.0$ |
|  | $>50 \sim 80$ | $1.5 \sim 2.4$ | $1.8-3.6$ | $1.9 \sim 4.6$ |
|  | $>80-120$ | $1.6 \sim 2.6$ | $2.0 \sim 3.9$ | 2. $1 \sim 5.2$ |
|  | $>120$ | $1.7 \sim 2.8$ | $2.2-4.2$ | $2.3-5.6$ |

解 a）载荷计算：

输人转矩 $T=9549 \times P / n$

$$
\begin{aligned}
& =9549 \times 8.83 / 1275 \mathrm{~N} \cdot \mathrm{m} \\
& =66.13 \mathrm{~N} \cdot \mathrm{m}
\end{aligned}
$$

名义切向力 $F_{1}=2000 \times T / d_{\mathrm{m}}$

$$
\begin{aligned}
& =2000 \times 66.13 /[(25+21) / 2] \mathrm{N} \\
& =5750.4 \mathrm{~N}
\end{aligned}
$$

单位载荷 $W=F_{\mathrm{l}} /(N l)$

$$
\begin{aligned}
& =5750.4 /(6 \times 30) \mathrm{N} / \mathrm{mm} \\
& =31.95 \mathrm{~N} / \mathrm{mm}
\end{aligned}
$$

b）齿面接触强度计算:

| 项 目 |  | 代号 | 式 |
| :---: | :---: | :---: | :---: |
| 戟荷计算 | 输人转矩 <br> 名义切向力 <br> 单位辌荷 | $T$ <br> $F_{1}$ <br> $W$ | $T=9549 P / n$ <br> $F_{\mathrm{t}}=2000 T / d_{\mathrm{m}}$ <br> $W=F_{1} /(N \cdot t)$ |
| 齿面接触 <br> 强度计算 | 先面压应力 <br> 齿面许用厓应力 <br> 满足条件 | $\left[\begin{array}{c}\sigma_{H} \\ {\left[\sigma_{H}\right]}\end{array}\right.$ | $\sigma_{H}=W / h_{W}$ <br> 式中: $h_{W}=h_{\min }$ <br> $\left[\sigma_{H}\right]=\sigma_{0,2} /\left(S_{H} K_{1} K_{2} K_{3} K_{4}\right)$ <br> $\sigma_{H} \leqslant\left[\sigma_{\mathrm{H}}\right]$ |
| ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-5.jpg?height=699&width=515&top_left_y=5123&top_left_x=645) | 齿根弯曲应力 <br> 许用弯曲应力 <br> 满足条件 | $\left[\left.\begin{array}{c}\sigma_{F} \\ {\left[\sigma_{F}\right]}\end{array} \right\rvert\,\right.$ | ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-5.jpg?height=699&width=3363&top_left_y=5123&top_left_x=2482) |
| 齿根抗等 <br> 强计算 | 齿根最大扭转 <br> 切应力 <br>  <br> 许用切应力 <br> 满足条件 | $\tau_{\text {Fmex }}$ <br>  <br> $\left[\tau_{\mathrm{F}}\right]$ | $\begin{aligned}$\tau_{F_{\text {max }}} & =\tau_{\mathrm{ma}} \alpha_{\mathrm{tm}} \\ $ 式中 $: \tau_{\mathrm{th}} & =\frac{16000 T}{\pi d_{\mathrm{b}}^{3}} \\ d_{\mathrm{h}} & =d+\frac{K d(D-d)}{D} \\ K &$ 值 $:$ 轻系列取 0.50 、中系列取 0.45 \\ $\alpha_{\mathrm{in}} & =\frac{d}{d_{\mathrm{h}}}\left(1+0.17 \frac{h}{\rho}\left(1+\frac{3.94}{0.1+\frac{h}{\rho}}\right)+\frac{6.38\left(1+0.1 \frac{h}{\rho}\right)}{\left(2.38+\frac{d}{2 h}\left(\frac{h}{\rho}+0.04\right)^{1 / 3}\right)^{2}}\right. \\ {\left[\tau_{\mathrm{F}}\right] } & =\left[\sigma_{F}\right] / 2 \\ \tau_{\mathrm{P}_{\mathrm{max}}} & \leqslant\left[\tau_{\mathrm{F}}\right]\end{aligned}$$ |
| ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-5.jpg?height=589&width=515&top_left_y=7438&top_left_x=645) | 茜面压应力 <br> 齿面擆许用应力 <br> 满足条件 | $\left[\begin{array}{c}\sigma_{H} \\ {\left[\sigma_{H 1}\right]}\end{array}\right]$ | $\sigma_{\mathrm{M}}=W / h_{W}$ <br> 式中 $: h_{W}=h_{\min }$ <br> 见表 $5.3-17$ <br> $\sigma_{H} \leqslant\left[\sigma_{H 1}\right]$ |
| 长期工作 <br> 无尿损 <br> 计算 | ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-5.jpg?height=625&width=865&top_left_y=8026&top_left_x=1251) | $\left[\begin{array}{c}\sigma_{\mathrm{H}} \\ \\ {\left[\sigma_{\mathrm{H} 2}\right]}\end{array}\right]$ | $\sigma_{\mathrm{H}}=W / h_{\mathrm{W}}$ <br> 式中: $h_{\mathrm{W}}=h_{\mathrm{min}}$ <br> 贝表 $5.3-17$ <br> $\sigma_{\mathrm{H}} \leqslant\left[\sigma_{\mathrm{H} 2}\right]$ |

荬面压应力 $\sigma_{H}=W / h_{n}$

$$
\begin{aligned}
& =31.95 / 2 \mathrm{MPa} \\
& =15.98 \mathrm{MPa}
\end{aligned}
$$

齿面许用压应力

$$
\begin{aligned}
{\left[\sigma_{\mathrm{H}}\right] } & =\sigma_{0.2} /\left(S_{\mathrm{H}} K_{1} K_{2} K_{3} K_{4}\right) \\
& =965 /(1.4 \times 1.25 \times 1.2 \times 1.3 \times 1.4) \mathrm{MPa} \\
& =252.5 \mathrm{MPa}
\end{aligned}
$$

襄 5.3-16 计算公式

|  | 项 目 | 代号 | 公 | 式 |
| :---: | :---: | :---: | :---: | :---: |
| 外花键扭 <br> 转与抗弯 <br> 强度计华 | 许用应力 | $\sigma_{v}$ <br>  <br> $\left[\sigma_{v}\right]$ | ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-6.jpg?height=1544&width=1856&top_left_y=1196&top_left_x=2327) |  |

取: $S_{\mathrm{H}}=1.4 、 K_{1}=1.25 、 K_{2}=1.2 、 K_{3}=1.3 、$ $K_{4}=1.4$

计算结果：满足 $\sigma_{H} \leqslant\left[\sigma_{H}\right]$ 条件，安全。

c) 齿根抗弯强度计算：

齿根弯曲应力 $\sigma_{\mathrm{F}}=6 \mathrm{hW} / \mathrm{S}_{\mathrm{Fa}}^{2}$

$$
\begin{aligned}
& =6 \times 2 \times 31.95 / 5^{2} \mathrm{MPa} \\
& =15.3 \mathrm{MPa}
\end{aligned}
$$

齿根许用弯曲应力

$\left[\sigma_{\mathrm{F}}\right]=\sigma_{\mathrm{b}} /\left(S_{\mathrm{F}} K_{1} K_{2} K_{3} K_{4}\right)$

$$
\begin{aligned}
& =1080 /(1.5 \times 1.25 \times 1.2 \times 1.3 \times 1.4) \mathrm{MPa} \\
& =263.7 \mathrm{MPa}
\end{aligned}
$$

取 $S_{\mathrm{P}}=1.5$

计算结果: 满足 $\sigma_{F} \leqslant\left[\sigma_{F}\right]$ 条件, 安全。

d）齿根抗槊强度计算：

齿根最大切应力

$$
\begin{aligned}
\tau_{\text {Frax }} & =\tau_{\mathrm{ta}} \alpha_{\mathrm{ta}} \\
& =29.5 \times 3.2 \mathrm{MPa} \\
& =94.4 \mathrm{MPa} \\
d_{\mathrm{b}} & =d+\frac{K d(D-d)}{D} \\
& =\left(21+\frac{0.45 \times 21(25-21)}{25}\right) \mathrm{mm} \\
& =22.51 \mathrm{~mm}
\end{aligned}
$$

上式中由表 5.3-18 查得 $K=0.45$

$$
\tau_{\mathrm{tn}}=\frac{16000 T}{\pi d_{\mathrm{b}}^{3}}
$$

$$
\begin{aligned}
& =\frac{16000 \times 66.13}{\pi \times 22.51^{3}} \mathrm{MPa} \\
& =29.5 \mathrm{MPa}
\end{aligned}
$$

| $\sigma_{\mathrm{HI}}$ 值 |  |  |  |  |  | $\sigma_{\mathrm{H} 2}$ 值 |  |
| :---: | :---: | :---: | :---: | :---: | :---: | :---: | :---: |
| 末经热 <br> 处理 <br> $20 \mathrm{HRC}$ | 调质 <br> 外理 <br> 28HRC | 摔 火 |  |  | 碳、城 <br> 淬火 <br> $60 \mathrm{HRC}$ | 末然热处理 | $0.028 \times$ 布氏唒度值 |
|  |  | 40HRC | 45 HRC | 5OHRC |  | 调质处理 | $0.032 \times$ 布氏硕度值 |
|  |  |  |  |  |  | 摔火 | $0.3 \times$ 洛氏硬度全 |
| 95 | 110 | 135 | 170 | 185 | 205 | ![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-6.jpg?height=139&width=717&top_left_y=7645&top_left_x=4085) | $0.4 \times$ 洛氏锗度值 |

$$
\begin{aligned}
\alpha_{\mathrm{la}}= & \frac{d}{d_{\mathrm{b}}}\left\{1+0.17 \frac{h}{\rho}\left(1+\frac{3.94}{0.1+\frac{h}{\rho}}\right)+\right. \\
& \left.\frac{6.38\left(1+0.1 \frac{h}{\rho}\right)}{\left(2.38+\frac{d}{2 h}\left(\frac{h}{\rho}+0.04\right)^{1 / 3}\right)^{2}}\right\} \\
= & \frac{21}{22.51}\left\{1+0.17 \times \frac{2}{0.2}\left(1+\frac{3.94}{0.1+\frac{2}{0.2}}\right)+\right.
\end{aligned}
$$

$$
\left.\frac{6.38\left(1+0.1 \times \frac{2}{0.2}\right)}{\left[2.38+\frac{21}{2 \times 2}\left(\frac{2}{0.2}+0.04\right)^{1 / 3}\right]^{2}}\right\}=3.2
$$

许用切应力 $\left[\tau_{\mathrm{F}}\right]=\left(\sigma_{\mathrm{F}}\right) / 2$

$$
\begin{aligned}
& =263.7 / 2 \mathrm{MPa} \\
& =131.9 \mathrm{MPa}
\end{aligned}
$$

计算结果: 满足 $\tau_{\text {Fmax }} \leqslant\left[\tau_{F}\right]$ 条件, 安全。

e）齿面㢦管损能力计算:

花键副在 $10^{8}$ 循环数下工作时耐度损能力计算:

齿面压应力 $\sigma_{\mathrm{H}}=15.98 \mathrm{MPa}$

齿面磨损许用压应力 $\left[\sigma_{\mathrm{H} 1}\right]=205 \mathrm{MPa}$ （㚗表 5. 3-17 得)

数 5.3-17 $\sigma_{\mathrm{mi}}$ 值、 $\sigma_{\mathrm{kz} 2}$ 值（摘自 GB/T 17855-1999）

计算结果: 满足 $\sigma_{\mathrm{H}} \leqslant\left[\sigma_{\mathrm{H} 1}\right]$ 条件, 安全。

花键副长期工作无唐损时耐箱损能力计算:

齿面压应力 $\sigma_{H}=22.8 \mathrm{MPa}$

啮面磨损许用压应力 $\left[\sigma_{\mathrm{H} 2}\right]=0.4 \times 58 \mathrm{MPa}=$ 23. $2 \mathrm{MPa}$ （査表 5.3-17 得）计算结果: 满足 $\sigma_{\mathrm{H}} \leqslant\left[\sigma_{\mathrm{H} 2}\right]$ 条件, 可以长期无黁损（或很少磨损）工作。

f) 外花键的规剪与抗弯强度计算：

当量应力 $\sigma_{v}=\sqrt{\sigma_{\mathrm{Pn}}^{2}+3 \tau_{m}^{2}}$

$$
=\sqrt{0^{2}+3 \times 29.5^{2}} \mathrm{MPa}
$$

$$
\begin{array}{rrr}
= & 51.1 \mathrm{MPa} \quad\left(M_{\mathrm{b}}=0 、 \sigma_{\mathrm{Pn}}=0\right) & 1.4) \mathrm{MPa} \\
\text { 许用应力 }\left[\sigma_{\mathrm{v}}\right]=\sigma_{0.2} /\left(S_{F} K_{1} K_{2} K_{3} K_{4}\right) & =235.7 \mathrm{MPa}
\end{array}
$$

$$
=965 /(1.5 \times 1.25 \times 1.2 \times 1.3 \times
$$

计算结果：满足 $\sigma_{v} \leqslant\left[\sigma_{v}\right]$ 条件，安全。

表 5.3-18 $K$ 值

| 轻系列矩形花键 | 0.5 | 较少齿渐开线花键 | 0.3 |
| :---: | :---: | :---: | :---: |
| 中系列矩形花键 | 0.45 | 较考齿海开线花键 | 0.15 |

## 2.3 矩形花键连接

2.3.1 矩形花键基本尺寸系列（见表 5.3-19、表 5.3-20）

| 小 <br> 径 <br> $d$ | 轻 系 列 |  |  |  |  | 中 系 列 |  |  |  |  |
| :---: | :---: | :---: | :---: | :---: | :---: | :---: | :---: | :---: | :---: | :---: |
|  | 规格 <br> $N \times d \times D \times B$ | c | $\boldsymbol{r}$ | 参考 |  | 规格 <br> $N \times d \times D \times B$ | c | $r$ | 參孝 |  |
|  |  |  |  | $d_{1 \text { min }}$ | $a_{\min }$ |  |  |  | $d_{1_{\text {min }}}$ | $a_{\text {mix }}$ |
| 11 <br> 13 |  |  |  |  |  | $6 \times 11 \times 14 \times 3$ <br> $6 \times 13 \times 16 \times 3.5$ | 0.2 | 0.1 |  |  |
| 16 |  |  |  |  |  | $6 \times 16 \times 20 \times 4$ |  |  | 14.4 | 1.0 |
| 18 |  |  |  |  |  | $6 \times 18 \times 22 \times 5$ | 0.3 | O? | 16.6 | 1.0 |
| 21 |  |  |  |  |  | $6 \times 21 \times 25 \times 5$ | 0.3 | 0.2 | 19.5 | 2.0 |
| 23 | $6 \times 23 \times 26 \times 6$ | 0.2 | 0.1 | 22 | 3.5 | $6 \times 23 \times 28 \times 6$ |  |  | 21.2 | 1.2 |
| 26 | $6 \times 26 \times 30 \times 6$ |  |  | 24.5 | 3.8 | $6 \times 26 \times 32 \times 6$ |  |  | 23.6 | 1.2 |
| 28 | $6 \times 28 \times 32 \times 7$ |  |  | 26.6 | 4.0 | $6 \times 28 \times 34 \times 7$ |  |  | 25.8 | 1.4 |
| 32 | $8 \times 32 \times 36 \times 6$ | 0.3 | 0.2 | 30.3 | 2.7 | $8 \times 32 \times 38 \times 6$ | 0.4 | 0.3 | 29.4 | 1.0 |
| 36 | $8 \times 36 \times 40 \times 7$ |  |  | 34.4 | 3.5 | $8 \times 36 \times 42 \times 7$ |  |  | 33.4 | 1.0 |
| 42 | $8 \times 42 \times 46 \times 8$ |  |  | 40.5 | 5.0 | $8 \times 42 \times 48 \times 8$ |  |  | 39.4 | 2.5 |
| 46 | $8 \times 46 \times 50 \times 9$ |  |  | 44.6 | 5.7 | $8 \times 46 \times 54 \times 9$ |  |  | 42.6 | 1.4 |
| 52 | $8 \times 52 \times 58 \times 10$ |  |  | 49.6 | 4.8 | $8 \times 52 \times 60 \times 10$ | 0.5 | 0.4 | 48.6 | 2.5 |
| 56 | $8 \times 56 \times 62 \times 10$ |  |  | 53.5 | 6.5 | $8 \times 56 \times 65 \times 10$ |  |  | 52.0 | 2.5 |
| 62 | $8 \times 62 \times 68 \times 12$ |  |  | 59.7 | 7.3 | $8 \times 62 \times 72 \times 12$ |  |  | 57.7 | 2.4 |
| 72 | $10 \times 72 \times 78 \times 12$ | 0.4 | 0.3 | 69.6 | 5.4 | $10 \times 72 \times 82 \times 12$ |  |  | 67.7 | 1.0 |
| 82 | $10 \times 82 \times 88 \times 12$ |  |  | 79.3 | 8.5 | $10 \times 82 \times 92 \times 12$ |  |  | 77.0 | 2.9 |
| 92 | $10 \times 92 \times 98 \times 11$ |  |  | 89.6 | 9.9 | $10 \times 92 \times 102 \times 11$ | 0.0 | 0.5 | 87.3 | 4.5 |
| 102 | $10 \times 102 \times 108 \times 16$ |  |  | 99.6 | 11.3 | $10 \times 102 \times 112 \times 16$ |  |  | 97.7 | 6.2 |
| 112 | $10 \times 112 \times 120 \times 18$ | 0.5 | 0.4 | 108.8 | 10.5 | $10 \times 112 \times 125 \times 18$ |  |  | 106.2 | 4.1 |

表 5.3-19 矩形花腱基本尺寸系列（摘自 GB/T1144-2001）

( $\mathrm{mm})$

![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-7.jpg?height=1194&width=2723&top_left_y=2671&top_left_x=610)

| 花键规格 | $N \times d \times D \times B \quad$ 例如 $6 \times 23 \times 26 \times 6$ |
| :--- | :---: |
| 花链 | $6 \times 23 \frac{\mathrm{H} 7}{\mathrm{f} 7} \times 26 \frac{\mathrm{H} 10}{\mathrm{al1}} \times 6 \frac{\mathrm{H} 11}{\mathrm{~d} 10} \quad \mathrm{~GB} / \mathrm{T} 1144-2001$ |
| 内花链 | $6 \times 23 \mathrm{H} 7 \times 26 \mathrm{H} 10 \times 6 \mathrm{H} 11 \quad \mathrm{~GB} / \mathrm{T} 1144-2001$ |
| 外花键 | $6 \times 23 \mathrm{H} \times 26 \mathrm{al1} \times 6 \mathrm{~d} 10 \quad \mathrm{~GB} / \mathrm{T} 1144-2001$ |

注：1.r一圆角半径； $D$ 一大径；B一键宽或链相宽。
2. $d_{1}$ 和 $a$ 值仅适用于展成法加工。

表 5.3-20 矩形内花钘型式及长度系列（摘自GB/T10081-2001）

( mm)
![](https://cdn.mathpix.com/cropped/2024_03_02_87261e08c2acfb0d9c31g-7.jpg?height=1078&width=2958&top_left_y=7424&top_left_x=1829)

```
