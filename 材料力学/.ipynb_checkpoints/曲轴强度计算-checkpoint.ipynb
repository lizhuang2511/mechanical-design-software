# 计算中间轴部分：计算的是中间轴的弯曲应力控制在100mpa以内，这个是将中间轴等效为简支梁计算的，

```python
import numpy as np
曲轴中间轴的宽y=150
中间轴直径d1=170
曲轴支壁肩厚度h=70
曲轴连接壁厚度h1=85#这个有异议
圆角半径r2=15
曲轴支撑轴直径d2=165
曲轴连接壁宽度b=190
能力载荷=800000
中间轴支点间隔l=2*曲轴支壁肩厚度h+曲轴中间轴的宽y+2*圆角半径r2
print('支点间隔l=',中间轴支点间隔l,'mm')
中间轴弯矩=(能力载荷*(2*中间轴支点间隔l-曲轴中间轴的宽y))/8
print('中间轴弯矩=',中间轴弯矩,'n*mm')
中间轴截面系数=np.pi*中间轴直径d1**3/32
print('中间轴截面系数=',中间轴截面系数)
中间轴的弯曲应力=中间轴弯矩/中间轴截面系数
print('中间轴的弯曲应力=',int(中间轴的弯曲应力),'n/mm2')#100mpa以下
```

# 曲轴连接壁部分计算：

载荷产生的弯曲应力要求在100mpa之内

```python
支点到曲轴连接壁中立轴的距离m=(曲轴支壁肩厚度h/2)+圆角半径r2
连接壁弯矩=能力载荷*支点到曲轴连接壁中立轴的距离m/2
print('连接壁弯矩=',连接壁弯矩)
曲轴支壁截面系数=曲轴支壁肩厚度h**2/6*曲轴连接壁宽度b
print('曲轴支壁截面系数=',曲轴支壁截面系数)
曲轴支壁壁弯曲应力=连接壁弯矩/曲轴支壁截面系数
print('曲轴支壁壁弯曲应力=',曲轴支壁壁弯曲应力)
曲轴连接壁截面系数1=曲轴连接壁厚度h1**2/6*曲轴连接壁宽度b
曲轴连接壁弯曲应力1=连接壁弯矩/曲轴连接壁截面系数1
print('曲轴连接壁弯曲应力1=',曲轴连接壁弯曲应力1)
```

旋转切向力产生的弯矩50mpa以下

```python
beita=0.042471
arf=0.346558
连杆力=能力载荷/np.cos(beita)
print(连杆力)
切向力=连杆力*np.sin(beita+arf)
print(切向力)
曲轴半径=75
扭矩=切向力*曲轴半径
print(扭矩/1000)
分力=连杆力*np.cos(beita+arf)
print(分力)
旋转弯矩=切向力/2*(曲轴半径-中间轴直径d1/2)#两个曲柄壁所以剪切力除以2
支壁肩截面系数=曲轴支壁肩厚度h*曲轴连接壁宽度b**2/6
连接壁截面系数=曲轴连接壁厚度h1*曲轴连接壁宽度b**2/6
支撑壁回转弯曲应力=旋转弯矩/支壁肩截面系数
连接壁回转弯曲应力=旋转弯矩/连接壁截面系数
print('支撑壁回转弯曲应力=',支撑壁回转弯曲应力)
print('连接壁回转弯曲应力=',连接壁回转弯曲应力)
```

合成应力控制在150mpa

```python
连接支壁合成应力=曲轴支壁壁弯曲应力+支撑壁回转弯曲应力
连接壁合成应力=连接壁回转弯曲应力+曲轴连接壁弯曲应力1
print('连接支壁合成应力=',连接支壁合成应力)
print('连接壁合成应力=',连接壁合成应力)
```

回转力
扭曲应力要求小于30mpa

```python
回转力产生的支壁扭曲应力=9*切向力*支点到曲轴连接壁中立轴的距离m/(4*曲轴连接壁宽度b*曲轴支壁肩厚度h**2)#两个曲柄壁所以剪切力为F/2
回转力产生的连接扭曲应力=9*切向力*支点到曲轴连接壁中立轴的距离m/(4*曲轴连接壁宽度b*曲轴连接壁厚度h1**2)#这里不是太清楚
print('回转力产生的支壁扭曲应力=',回转力产生的支壁扭曲应力)
print('回转力产生的连接扭曲应力=',回转力产生的连接扭曲应力)
```

弯扭复合应力控制在120mpa

```python
支壁肩弯扭复合=0.35*曲轴支壁壁弯曲应力+0.65*np.sqrt(曲轴支壁壁弯曲应力**2+4*回转力产生的支壁扭曲应力**2)#这里有些问题要用弯曲复合计算
连接壁弯扭复合=0.35*曲轴连接壁弯曲应力1+0.65*np.sqrt(曲轴连接壁弯曲应力1**2+4*回转力产生的连接扭曲应力**2)#这里有些问题
print('支壁肩弯扭复合=',支壁肩弯扭复合)
print('连接壁弯扭复合=',连接壁弯扭复合)
```

# 曲轴支撑部分

扭曲应力在70mpa以下

```python
支撑部分扭矩应力=16*切向力*曲轴半径/(np.pi*曲轴支撑轴直径d2**3)
print('支撑部分扭矩=',支撑部分扭矩)
```

p和旋转力产生的弯曲应力和合应力，要求在120mpa以下

```python
支撑轴弯矩=曲轴支撑轴直径d2/2*np.sqrt((能力载荷/2)**2+(切向力/2)**2)#这个对吗
支撑轴截面系数=np.pi*曲轴支撑轴直径d2**3/32
支撑轴应力总=支撑轴弯矩/支撑轴截面系数#100mpa以下
支撑轴弯扭合成应力=0.35*支撑轴应力总+0.65*np.sqrt(支撑轴应力总**2+4*支撑部分扭矩应力**2)#这里有些问题
print('支撑轴弯扭合成应力=',支撑轴弯扭合成应力)
```
