平面应力状态任意平面应力表达公式

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
def 平面应力状态(oox,ooy,th,ttxy):
    ooth=(oox+ooy)/2+(oox-ooy)/2*np.cos(2*th)-ttxy*np.sin(2*th)
    ttth=(oox-ooy)/2*np.sin(2*th)+ttxy*np.cos(2*th)
    return ooth,ttth
```

```python
oox=100
th=np.linspace(0,2*np.pi,360)
平面应力状态0=平面应力状态(oox,0,th,0)
plt.plot(np.rad2deg(th),平面应力状态0[0])
plt.plot(np.rad2deg(th),平面应力状态0[1])
```
