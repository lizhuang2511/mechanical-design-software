已知参数：
${{\theta }_{0}},r,{{l}_{1}},{{l}_{2}},{{l}_{3}},{{l}_{4}},{{l}_{5}},{{x}_{3}},{{y}_{3}},N$ ，spm，其中theta0数组使用numpy库定义为曲柄转角0-360度，${{\theta }_{0}}$在数组中取值去计算每个角度的多连杆位置


```python
import numpy as np
r=192.5
l1=1605
l2=900
l3=920
l4=850
l5=930
x3=2130
y3=350
n=20
x5=x3+n
spm=20
theta0数组=np.linspace(0, 2*np.pi, 1000)#0到360度分为1000份
theta0=theta0数组[0]
```

定义多连杆运动方程：$\begin{align}
  & \text{r}\times \cos ({{\theta }_{0}})+{{l}_{1}}\times \cos ({{\theta }_{1}})+{{l}_{2}}\times \cos ({{\theta }_{2}})={{x}_{3}}-0 \\ 
 & \text{r}\times \sin ({{\theta }_{0}})+{{l}_{1}}\times \sin ({{\theta }_{1}})+{{l}_{2}}\times \sin ({{\theta }_{2}})={{y}_{3}}-0 \\ 
 & {{l}_{2}}\times \cos ({{\theta }_{2}})+{{l}_{3}}\times \cos ({{\theta }_{3}})={{l}_{4}}\times \cos ({{\theta }_{4}}) \\ 
 & {{l}_{2}}\times \sin ({{\theta }_{2}})+{{l}_{3}}\times \sin ({{\theta }_{3}})={{l}_{4}}\times \sin ({{\theta }_{4}}) \\ 
 & {{l}_{3}}\times \cos ({{\theta }_{3}})+{{l}_{5}}\times \cos ({{\theta }_{5}})={{x}_{5}}-{{x}_{3}} \\ 
 & {{l}_{3}}\times \sin ({{\theta }_{3}})+{{l}_{5}}\times \sin ({{\theta }_{5}})={{y}_{5}}-{{y}_{3}} \\ 
\end{align}$

未知数：${{\theta }_{1}},{{\theta }_{2}},{{\theta }_{3}},{{\theta }_{4}},{{\theta }_{5}},{{y}_{5}}$

```python
# 定义方程组的函数
def equations(unknowns):
    theta1, theta2, theta3, theta4, theta5, y5 = unknowns
    # 方程1
    eq1 = r * np.cos(theta0) + l1 * np.cos(theta1) + l2 * np.cos(theta2) - x3
    
    # 方程2
    eq2 = r * np.sin(theta0) + l1 * np.sin(theta1) + l2 * np.sin(theta2) - y3
    
    # 方程3
    eq3 = l2 * np.cos(theta2) + l3 * np.cos(theta3) - l4 * np.cos(theta4)
    
    # 方程4
    eq4 = l2 * np.sin(theta2) + l3 * np.sin(theta3) - l4 * np.sin(theta4)
    
    # 方程5
    eq5 = l3 * np.cos(theta3) + l5 * np.cos(theta5) - (x5 - x3)
    
    # 方程6
    eq6 = l3 * np.sin(theta3) + l5 * np.sin(theta5) - (y5 - y3)
    
    return [eq1, eq2, eq3, eq4, eq5, eq6]
```

猜测未知数初始值：${{\theta }_{1}},{{\theta }_{2}},{{\theta }_{3}},{{\theta }_{4}},{{\theta }_{5}},{{y}_{5}}$

```python
# 初始猜测值（这些值应该根据问题的具体情况进行合理选择）
initial_guess = [np.deg2rad(360),np.deg2rad(70),np.deg2rad(270),np.deg2rad(0), np.deg2rad(270),-l3-l5]
```

使用scipy包中牛顿-辛普森方法解非线性方程，解出未知数

```python
from scipy.optimize import fsolve
# 使用fsolve求解方程组
solution = fsolve(equations, initial_guess)
# 提取解
theta1_sol, theta2_sol, theta3_sol, theta4_sol, theta5_sol, y5_sol = solution
initial_guess=solution
print("theta1_sol,theta2_sol,theta3_sol,theta4_sol,theta5_sol,y5_sol=",
      theta1_sol, theta2_sol, theta3_sol, theta4_sol, theta5_sol, y5_sol)
```

360度循环求解，每次解出结果为下次猜测值，使用pandas库保存滑块位移数据，对时间求导计算滑块速度，加速度

```python
import pandas as pd
角速度=spm*2*np.pi/60#弧度/s
时间=np.linspace(0,2*np.pi/角速度,1000)#时间
数据=pd.DataFrame()
滑块y坐标=[]
for i in range(len(theta0数组)):
    theta0=theta0数组[i]
    initial_guess=solution
    solution=fsolve(equations, initial_guess)
    theta1_sol, theta2_sol, theta3_sol, theta4_sol, theta5_sol, y5_sol = solution
    滑块y坐标.append(y5_sol)
滑块速度=np.gradient(滑块y坐标,时间,edge_order=2)
滑块加速度=np.gradient(滑块速度,时间,edge_order=2)
距离下死点位置=滑块y坐标-min(滑块y坐标)
数据['角度']=theta0数组/np.pi*180
数据['距离下死点位置']=距离下死点位置
数据['滑块速度']=滑块速度
数据['滑块加速度']=滑块加速度
```

处理数据将下死点设置为180度

```python
行数 = len(数据)
最小行程位置 = 数据.loc[数据.loc[:, '距离下死点位置'] == min(数据.loc[:, '距离下死点位置']), :].index[0]
补差行数=abs(int(行数/2-最小行程位置))
if 最小行程位置>=行数/2:#需要前补后
    数据调整1 = 数据.loc[补差行数:]
    数据调整2 = 数据.loc[:补差行数-1]
    数据 = pd.concat([数据调整1, 数据调整2], ignore_index=True)
if 最小行程位置<行数/2:#需要后补前
    补差=abs(int(行数-补差行数))
    数据调整1 = 数据.loc[补差:]
    数据调整2 = 数据.loc[:补差 - 1]
    数据 = pd.concat([数据调整1, 数据调整2], ignore_index=True)
角度重置 = np.linspace(0, 2 * np.pi, 行数)
数据['角度'] = 角度重置 / np.pi * 180
数据 = 数据.reset_index(drop=True)
```

使用matplotlib库画图展示数据

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
from matplotlib.pyplot import MultipleLocator
DATA=数据
最大位移=max(数据['距离下死点位置'])
fig, ax = plt.subplots()
ax.set_title("角度-滑块位置曲线")
ax.set_xlabel("角度（°）")
ax.set_ylabel("距离下死点位置(mm)")
ax.set_xlim(0, 360)
ax.set_ylim(min(DATA['距离下死点位置'].replace([np.inf, -np.inf], np.nan).dropna()) - abs(
    min(DATA['距离下死点位置'].replace([np.inf, -np.inf], np.nan).dropna()) / 10),
            max(DATA['距离下死点位置'].replace([np.inf, -np.inf], np.nan).dropna()) + abs(
                max(DATA['距离下死点位置'].replace([np.inf, -np.inf], np.nan).dropna()) / 10))
# - PLOT ------------------
ax.plot(DATA["角度"].replace([np.inf, -np.inf], np.nan), DATA["距离下死点位置"].replace([np.inf, -np.inf], np.nan),
        linewidth=1.0, alpha=1.0, color="#000000")
x_major_locator = MultipleLocator(20)
# 把x轴的刻度间隔设置为1，并存在变量里
if int(最大位移) - ax.axis()[2] <= 100:
    y_major_locator = MultipleLocator(round(int((int(最大位移) - ax.axis()[2]) / 10)))
else:
    y_major_locator = MultipleLocator(round(int((int(最大位移) - ax.axis()[2]) / 10)))
# 把y轴的刻度间隔设置为10，并存在变量里
ax = plt.gca()
# ax为两条坐标轴的实例
ax.xaxis.set_major_locator(x_major_locator)
# 把x轴的主刻度设置为1的倍数
ax.yaxis.set_major_locator(y_major_locator)
plt.grid(linestyle='--')
# plt.text(最大位移发生的角度-1, 最大位移+1, "行程:"+str(int(最大位移))+"mm", size = 10)
plt.legend(labels=["行程:" + str(int(最大位移)) + "mm"], loc="lower right", fontsize=10)
# ax.axhline(最大位移,xmin=0,xmax=(最大位移-ax.axis()[0])/(ax.axis()[1]-ax.axis()[0]),linestyle='--')
plt.show()
```
